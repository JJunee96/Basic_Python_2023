### 인터넷 접속 라이브러리 추가

- urllib.request 모듈로 작업
- html 소스 로드로 별다른 문제없음

````python
from urllib.request import urlopen, Request

# 도시별 날씨 검색함수
def get_weather(city):
    # 기상청 홈페이지 도시별 날씨 페이지
    url = 'https://www.weather.go.kr/w/obs-climate/land/city-obs.do'
    page = urlopen(url=url)

    text = page.read().decode('utf-8')
    text =text[text.find(f'>{city}</a>'):]

    # 기온 가져오기 반복문 13번
    for i in range(7):
        text = text[text.find('<td>') + 1:]   # text 내용 가져오기

    start = 3   # 3번째 내용 시작하는부분
    end = text.find('</td>')    # 내용 끝나는부분
    current_temp = text[start:end] # 내용 부분만 짜르기
    print(f'{city}의 현재 기온은 {current_temp}˚C 입니다.')

    # 습도 가져오기
    for i in range(3):
        text = text[text.find('<td>') + 1:]

    start = 3
    end = text.find('</td>')
    current_humid = text[start:end]
    print(f'{city}의 현재 습도는 {current_humid}% 입니다.')


get_weather('부산')
````

- 문제점
    - https를 일반 request로 부르면 SSL 오류발생
    - 해결하려면 외부모듈 requests를 사용, verify = False 옵션을 추가로 지정

````python
import requests
from urllib.parse import quote, unquote, urlencode  # 한글을 URLencode 변환하는 함수    ex) 롯데 --> '%EB%A1%AF%EB%8D%B0'
import json
import ssl

# stationName 정류장이름
def getDataPortalSearch(stationName, type):
    api_url = 'https://apis.data.go.kr/6260000/BusanTblBusinfoeqStusService/getTblBusinfoeqStusInfo'
    queryString = "?" + urlencode(
        {
            'serviceKey':'YYHJjeiagyFeSD69gRR9ykzP7KS0HqJOPMzv0HmJ8ETzz2witGrEIA63UsK40QjbpFfTSI2lRJ23P3/sHnjoRw==',
            'pageNo':'1',
            'numOfRows' : '10',
            'resultType': type,
            'stationLoc': stationName
        }
    )

    total_url = api_url + queryString
    response = requests.get(total_url, verify=False)
    return response.text

try:
    result = getDataPortalSearch('백화점', 'json')
    json_data = json.loads(result)
    station_data = json_data['getTblBusinfoeqStusInfo']['body']['items']['item']    # 필요한 내용만 뽑아오기

    for i in station_data:  # 리스트 대로 나열
        print(i)
except Exception as e:
    print('찾는 데이터가 없습니다.')
````

### OpenAPI로 가져온 데이터 지도 표시

```python
stop_str =  '<h4>' + item['stationLoc'] + '</h4>' + item['stationNum'] + '<br>' + item['addr']
```
- 문제점
    - 딕셔너리 구조하고 문자열 포맷팅 f'{}' 호환 안됨!
    - 구식방법인 문자열 결합방식으로 해결

````python
import folium

if len(station_data) > 0:   # 정류장 중 제일 첫번째 인덱스 정류장 위경도를 중심으로
    center_lat = station_data[0]['lat'] # 위도 데이터
    center_lng = station_data[0]['lng'] # 경도 데이터
    

m = folium.Map(location = [ center_lat,center_lng], zoom_start=13)

# 전체 정류장 위치값 마커 표시
for item in station_data:
    stop_str =  '<h4>' + item['stationLoc'] + '</h4>' + item['stationNum'] + '<br>' + item['addr']
    iframe = folium.IFrame(stop_str)
    popup = folium.Popup(iframe, min_width = 200, max_width = 200) # 팝업 사이즈 지정
    folium.Marker(location=[item['lat'],item['lng']], popup= popup
                , icon = folium.Icon(icon = 'pushpin')).add_to(m)

m
````

````python
from bs4 import BeautifulSoup
import requests

url = 'https://kin.naver.com/search/list.naver?query=%EC%A0%9C%EC%9E%84%EC%8A%A4+%EC%9B%B9+%EB%A7%9D%EC%9B%90%EA%B2%BD'

response = requests.get(url)

if response.status_code == 200: # 웹페이지 OK
    html = response.text
    soup = BeautifulSoup(html, 'html.parser') # parser : 분석하는 도구
    # print(soup)
    title = soup.select_one('#s_content > div.section > ul > li:nth-child(1) > dl > dt > a')
    print(title.get_text())
else:
    print(f'Error : {response.status_code}')
````

````python
from bs4 import BeautifulSoup
import requests

url = 'https://kin.naver.com/search/list.naver?query=%EC%A0%9C%EC%9E%84%EC%8A%A4+%EC%9B%B9+%EB%A7%9D%EC%9B%90%EA%B2%BD'

response = requests.get(url)

if response.status_code == 200: # 웹페이지 OK
    html = response.text
    soup = BeautifulSoup(html, 'html.parser') # parser : 분석하는 도구
    ul = soup.select_one('ul.basic1')
    titles = ul.select('li > dl > dt > a')
    for title in titles:
        print(title.get_text())
else:
    print(f'Error : {response.status_code}')
````
